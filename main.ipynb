```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
def mb_pdf(x, beta):
    """Define the Maxwell-Boltzmann Distribution"""
    
    arg = -(x**2) / (2*beta**2)
    prefactor = np.sqrt(2/np.pi) * x**2 / beta**3
    func = prefactor * np.exp(arg)

    return func
```

```python
x = np.arange(0,20,0.1)

#Values of 
betas = np.arange(1, 5, 1)

for beta in betas:
    y = mb_pdf(x, beta)
    plt.plot(x,y, label=f'beta={beta}')

plt.title('Maxwell-Boltzmann PDF For Different Temperatures')
plt.ylabel('PDF')
plt.xlabel('Speed')
plt.legend()
plt.show()
```

```python
def mb_cdf(x, beta):
    """Define the CDF of the MB distribution"""

    cdf_vals = []

    for x_val in x:
        cdf_val = integrate.quad(mb_pdf, min(x), x_val, args = (beta))
        cdf_vals.append(cdf_val)

    return cdf_vals
```

```python
for beta in betas:
    y = mb_cdf(x, beta)
    plt.plot(x,y, label=f'beta={beta}')

plt.title('Maxwell-Boltzmann CDF For Different Temperatures')
plt.ylabel('CDF')
plt.xlabel('Speed')
plt.legend()
plt.show()
```
